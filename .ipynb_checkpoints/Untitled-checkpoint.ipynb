```python
import os
call='.item-'
url='background:url('
opts=['background-repeat: no-repeat;','background-position: center;','background-size: contain;']
prefix='\'/about/portfolio/images/desks/'
```

```python
images=os.listdir('about/portfolio/images/desks')
```

```python
for i in range(1,len(images)):
    print(call+str(i)+'{')
    print(str(url)+str(prefix)+str(images[i])+'\');')
    for j in opts:
        print(j)
    print('}')
```

```python
with open('test_carousel.css','w') as f:
    for i in range(1,len(images)):
        f.writelines(call+str(i)+'{\n')
        f.writelines(str(url)+str(prefix)+str(images[i])+'\');\n')
        for j in opts:
            f.writelines(j+str('\n'))
        f.writelines('}\n')
```
